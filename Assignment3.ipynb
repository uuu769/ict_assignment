```python
def find_max_builtin(a, b, c):
  return max(a, b, c)
print(find_max_builtin(10, 5, 20))
```

```python
def sum_list_builtin(numbers):
  return sum(numbers)

my_list = [18, 12, 30, 10, 7]
total = sum_list_builtin(my_list)
print(f"The sum of {my_list} is: {total}")
```

```python
my_list = [10, 20, 40, 50]
print(f"Original List: {my_list}")
number_to_insert = int(input("Enter the number you want to insert: "))
position = int(input(f"Enter the position (0-{len(my_list)}) to insert the number: "))
my_list.insert(position, number_to_insert)
print(f"Updated List: {my_list}")
```

```python
sample_dict = { "name": "Kelly", "age": 25, "salary": 8000, "city": "New york"}
city_value = sample_dict.pop("city")
sample_dict["location"] = city_value

print(sample_dict)
```
